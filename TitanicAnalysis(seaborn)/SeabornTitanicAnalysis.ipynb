# Seaborn 



## The Data

We will be working with a famous titanic data set for these exercises.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

# Analysis



```python
sns.jointplot(x='fare',y='age',data=titanic,xlim=(-100,600),ylim=(0,80))
```

```python
sns.displot(titanic['fare'],bins=20,color='lightcoral',lw=0,ls='-')
```

```python
figure=plt.figure(figsize=(6,5))
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

```python

sns.countplot(x='sex',data=titanic)
```

```python
df=titanic.corr()
sns.heatmap(data=df,cmap='coolwarm')
```

```python
g=sns.FacetGrid(data=titanic,col='sex')
g.map(sns.histplot,x='age',data=titanic)
```
